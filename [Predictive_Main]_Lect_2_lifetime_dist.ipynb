# Properties of commonly used lifetime distributions
## Exponential distribution
* Basic information: 
 * Only one parameter: $\lambda$ (failure rate).
 * Probability Density Function (PDF): $f(t) = \lambda {\rm e}^{-\lambda t}, t\ge 0.$
 * Cumulative Density Function (CDF): $F(t) = 1 - {\rm e}^{-\lambda t}, t\ge 0.$
 * Reliability function (Inverse CDF): $R(t) = 1-F(t) = {\rm e}^{-\lambda t}, t\ge 0.$
 * Failure rate (hazard rate): $\lambda.$ 
 * MTTF: $\frac{1}{\lambda}.$
* Important properties:
 * Positive supports.
 * PDF is a decreasing function, with peak at $t=0.$
 * The only distribution with constant failure rate.
 * Widely-used to model the failure of electronic products.
* Examples:
 * Suppose the lifetime of a product follows an exponential distribution with a failure rate $\lambda = 10^{-3}$ (year$^{-1}$).
 * Draw the PDF, reliability function, and failure rate function of this product. Verify the above important properties hold.

```python
from reliability.Distributions import Exponential_Distribution
import matplotlib.pyplot as plt
# Create an object for the exponentially distributed lifetime.
failure_rate = 0.001
exp_dist = Exponential_Distribution(Lambda=failure_rate)

# Call the PDF() method to draw PDF.
exp_dist.PDF()
plt.show()
# Call the SF() method to draw reliability function (SF: Survival Function).
exp_dist.SF()
plt.show()
# Call the HF() method to draw failure rate function (HF: Hazard Function).
exp_dist.HF()
plt.show()
```

## Weibull distribution
* Basic information: 
 * Two parameters:
   * Scale parameter (chracteristic life): $\alpha.$
   * Shape parameter: $\beta.$
 * Probability Density Function (PDF): $ f(t) = \frac{\beta t^{ \beta - 1}}{ \alpha^ \beta} {\rm e}^{-(\frac{t}{\alpha })^ \beta}.$
 * Cumulative Density Function (CDF): $F(t) = 1 - {\rm e}^{-(\frac{t}{\alpha })^ \beta }.$
 * Reliability function (Inverse CDF): $R(t) = {\rm e}^{-(\frac{t}{\alpha })^ \beta }.$
 * Failure rate (hazard rate): $\frac{\beta}{\alpha} (\frac{t}{\alpha})^{\beta -1}.$ 
* Important properties:
 * Positive supports ($t\ge 0$).

 * Increasing? Constant? Decreasing failure rate?
* Examples:
 * Suppose the lifetime of a product follows a Weibull distribution with $\alpha = 1$ (year) and $\beta = 0.8.$

```python
from reliability.Distributions import Weibull_Distribution
import matplotlib.pyplot as plt
# Create an object for the Weibull distribution.
alpha = 1
beta = .8
wbl_dist = Weibull_Distribution(alpha, beta)

# Call the PDF() method to draw PDF.
wbl_dist.PDF()
plt.show()
# Call the SF() method to draw reliability function (SF: Survival Function).
wbl_dist.SF()
plt.show()
# Call the HF() method to draw failure rate function (HF: Hazard Function).
wbl_dist.HF()
plt.show()
```

## Lognormal distribution
* Basic information: 
 * Two parameters:
   * Scale parameter : $\mu.$
   * Shape parameter: $\sigma.$
 * Probability Density Function (PDF): $ f(t) = \frac{1}{\sigma t \sqrt{2\pi}} {\rm exp} \left[-\frac{1}{2} \left(\frac{{\rm ln}(t)-\mu}{\sigma}\right)^2\right], $ where $\phi$ is the standard normal PDF with $\mu = 0$ and $\sigma=1.$
 * Cumulative Density Function (CDF): $F(t) = \Phi \left( \frac{{\rm ln}(t) - \mu}{\sigma} \right), $ where $\Phi$ is the standard normal CDF with $\mu = 0$ and $\sigma=1.$
 * Reliability function (Inverse CDF): $R(t) = 1 - \Phi \left( \frac{{\rm ln}(t) - \mu}{\sigma} \right).$
 * Failure rate (hazard rate): $h(t) = \frac{\phi \left[ \frac{{\rm ln}(t) - \mu}{\sigma} \right]}{t \sigma \left(1 - \Phi \left( \frac{{\rm ln}(t) - \mu}{\sigma} \right)\right)}.$ 
* Important properties:
 * Positive supports ($t\ge 0$).
 * Shape of the PDF?
 * Increasing? Constant? Decreasing failure rate?
* Examples:
 * Suppose the lifetime of a product follows a Lognormal distribution with $\mu = 1$ (year) and $\sigma = 0.3.$

   * The shape of the PDF.

```python
from reliability.Distributions import Lognormal_Distribution
import matplotlib.pyplot as plt
# Create an object for the Weibull distribution.
mu = 1
sigma = .3
lognormal_dist = Lognormal_Distribution(mu, sigma)

# Call the PDF() method to draw PDF.
lognormal_dist.PDF()
plt.show()
# Call the SF() method to draw reliability function (SF: Survival Function).
lognormal_dist.SF()
plt.show()
# Call the HF() method to draw failure rate function (HF: Hazard Function).
lognormal_dist.HF()
plt.show()
```

# Combination of the distributions.
## Mixture of distributions.
In a lot of practical cases, the total population could be viewed as a mixture of two or several sub-populations. For example, in manufacturing process, the process might subject to some unknown noise. As a result, the parts manufactured in a specific time interval might be significantly different from the normal output of the process. This phenomenon could be modeled by mixture of distributions.

* Let us assume the total population contains two sub-populations. 
* Let $f(t)$ denote the PDF of the entire population, 
* and let $f_1(t), f_2(t)$ denote the PDF of the first and second sub-population, respectively. 
* Let us assume the a fraction of $p$ among the entire population belongs to sub-population $1,$ and $(1-p)$ of the total population belongs to the second sub-population.

Then, we can easily derive: $f(t) = p f_1(t) + (1-p) f_2(t).$ The other reliabilty metrics can be derived accordingly.


```python
from reliability.Distributions import Lognormal_Distribution, Gamma_Distribution, Weibull_Distribution, Mixture_Model
import matplotlib.pyplot as plt

# create the mixture model
d1 = Lognormal_Distribution(mu=2, sigma=0.8)
d2 = Weibull_Distribution(alpha=50, beta=5, gamma=100)
d3 = Gamma_Distribution(alpha=5, beta=3, gamma=30)
mixture_model = Mixture_Model(distributions=[d1, d2, d3], proportions=[0.3, 0.4, 0.3])

# plot the 5 functions using the plot() function
mixture_model.plot()

# plot the PDF and CDF
plot_components = True # this plots the component distributions. Default is False
plt.figure(figsize=(9, 5))
plt.subplot(121)
mixture_model.PDF(plot_components=plot_components, color='red', linestyle='--')
plt.subplot(122)
mixture_model.CDF(plot_components=plot_components, color='red', linestyle='--')
plt.subplots_adjust(left=0.1, right=0.95)
plt.show()

# extract the mean of the distribution
print('The mean of the distribution is:', mixture_model.mean)

'''
The mean of the distribution is: 74.91607709895453
'''
```

* Important feature: mixture of distributions might have multiple peaks. A lot of practical data acutally have this property.

## Competing risk models.
Another commonly encoutered phenomenon is that, an item might fail due to two differnet failure modes. Let us assume the first failure mode has reliabilty function $R_1(t)$ and the second $R_2(t)$. Let us further assume that the two failure modes are independent. Then, the reliability of the item could be determined from the competing risk model:
$R(t) = R_1(t) \cdot R_2(t). $

The other reliability metrics could be determined accordingly.



```python
from reliability.Distributions import Lognormal_Distribution, Gamma_Distribution, Weibull_Distribution, Competing_Risks_Model
import matplotlib.pyplot as plt

# create the competing risks model
d1 = Lognormal_Distribution(mu=4, sigma=0.1)
d2 = Weibull_Distribution(alpha=50, beta=2)
d3 = Gamma_Distribution(alpha=30,beta=1.5)
CR_model = Competing_Risks_Model(distributions=[d1, d2, d3])

# plot the 5 functions using the plot() function
CR_model.plot()

# plot the PDF and CDF
plot_components = True # this plots the component distributions. Default is False
plt.figure(figsize=(9, 5))
plt.subplot(121)
CR_model.PDF(plot_components=plot_components, color='red', linestyle='--')
plt.subplot(122)
CR_model.CDF(plot_components=plot_components, color='red', linestyle='--')
plt.show()
```
