# MLE for complete data
Suppose we have the following time to failure data:

10, 8, 12, 12, 15, 20, 8, 21, 9, 23, 21, 45, 30, 23, 52, 26, 24, 34, 50, 21, 23, 67, 46, 34, 34, 21, 23, 45, 51, 18, 23, 34, 22, 21, 30, 34, 28, 10, 40, 34

Use the MLE to estimate the parameters of the lifetime distribution.

## Fit_Everything()
See the document here: https://reliability.readthedocs.io/en/latest/API/Fitters/Fit_Everything.html#

```python
from reliability.Fitters import Fit_Everything
import matplotlib.pyplot as plt

# Define the data.
data = [10, 8, 12, 12, 15, 20, 8, 21, 9, 23, 21, 45, 30, 
        23, 52, 26, 24, 34, 50, 21, 23, 67, 46, 34, 34, 
        21, 23, 45, 51, 18, 23, 34, 22, 21, 30, 34, 28, 10, 40, 34]
# MLE on the Weibull distribution.
Fit_Everything(failures=data)
plt.show()
```

## Model selection metrics:
* Loglikelihood: The larger, the better.
* Akaike information criterion (AIC): 
  * $AIC = 2k - 2ln(L_{max})$
    * $k$: Number of parameters in your model
    * $L_{max}$: The maximized likelihood.
  * AIC rewards maximized likelihood while penalizing more model parameters.
  * The smaller, the better.
* Bayesian information criterion (BIC):
  * $BIC = kln(n) - 2ln(L_{max})$
    * $k$: Number of parameters in your model
    * $n$: Sample size of the observations.
    * $L_{max}$: The maximized likelihood.
    * The smaller, the better.
